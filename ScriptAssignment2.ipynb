## 8. Manipulating Data (Use: the functions below) - 1 point

```python
#The question was incomplete so I was unsure what to do for it, so I added some functions that manipulate data

import pandas as pd

dFrame = pd.read_csv('educ_figdp_1_Data.csv')

#replace : with NaN
dFrame['Value'] = dFrame['Value'].replace(':', pd.NA)

#convert the column to integers
dFrame['Value'] = pd.to_numeric(dFrame['Value'], errors='coerce')

#group data by column and calculate the mean of the Value column
groupData = dFrame.groupby('GEO')['Value'].mean()
print('Grouped data: \n', groupData)

#create a pivot table
pt = dFrame.pivot_table(values = 'Value', index = 'GEO', columns = 'TIME', aggfunc = 'sum')
print('\nPivot Table: \n', pt)

#unpivot the dataframe
meltData = pd.melt(dFrame, id_vars = ['GEO'], value_vars = ['Value'])
print('\nUnpivoted dataframe: \n', meltData)
```

## 11. Rearrange, rank and visualize the data (Use: Pivot Table, rank, stacked bar chart) - Bonus - 5 point
- Rearrange 1 points
- Rank 1 point
- Visualiation 3 points

```python
import pandas as pd
import matplotlib.pyplot as plt

dFrame = pd.read_csv('educ_figdp_1_Data.csv')

#replace : with NaN
dFrame['Value'] = dFrame['Value'].replace(':', pd.NA)

#convert the column to integers
dFrame['Value'] = pd.to_numeric(dFrame['Value'], errors = 'coerce')

#create a pivot table
pt = dFrame.pivot_table(values = 'Value', index = 'GEO', columns = 'TIME', aggfunc = 'sum')
print('\nPivot Table: \n', pt)

#rank pivot table data for each year (TIME)
rankedPT = pt.rank(ascending = False)
print('\nRanked Pivot Table: \n', rankedPT)

#visualize pivot table data using a stacked bar chart
pt.fillna(0, inplace = True) #replace NaN with 0 for visualization
pt.plot(kind = 'bar', stacked = True, figsize = (10, 7))

plt.title('Stacked Bar Chart of Value by GEO and TIME')
plt.xlabel('GEO (Country)')
plt.ylabel('Value')
plt.legend(title='Year')
plt.tight_layout()

plt.show()

#print ranked pivot table
print('Ranked Pivot Table: \n', rankedPT)
```
