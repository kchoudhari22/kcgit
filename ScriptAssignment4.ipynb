```python
import pandas as pd
import numpy as np

#load the data into a dataframe
macroDF = pd.read_csv('macrodata.csv')

print(macroDF)
```

Use unstack on years 1959-1999 and then stack for the remaining years in the years column. - 2 points

```python
#split the years into two categories 1959-1999 and 1999 and after and apply unstack/stack
unstackData = macroDF[(macroDF['year'] >= 1959) & (macroDF['year'] <= 1999)].set_index(['year', 'quarter']).unstack()
stackData = macroDF[macroDF['year'] > 1999].set_index(['year', 'quarter']).stack()

print('\n Unstacked Years: \n', unstackData)
print('\n Stacked Years: \n', stackData)
```

Use Swaplevel on columns "cpi" and "realint" - 1 Point

```python
#copy columns cpi and realint from dataframe
macroDF1 = macroDF[['cpi', 'realint']].copy()

#make a multiindex with two levels that groups columns under Factors
macroDF1.columns = pd.MultiIndex.from_tuples([('Factors', 'cpi'), ('Factors', 'realint')])

#swap the two levels by columns
swaplevel = macroDF1.swaplevel(axis = 1)

print('\n MultiIndex: \n', macroDF1)
print('\n Swaplevel: \n', swaplevel)
```

Merge all columns that start with "real"  into new column "unreal". (Use RE from the previous notebook if needed) - 1 Point

```python
#get all columns that start with real
real = macroDF.filter(regex = '^real').columns

#merge all real columns values into unreal column
macroDF['unreal'] = macroDF[real].sum(axis = 1)

print('\n Unreal column: \n', macroDF.unreal)
print('\n Unreal and Real columns: \n', macroDF[['unreal'] + list(real)])
```

Provide an example for join() - 1 Point

```python
#split the dataframe to demonstrate how to use join
macroDF2 = macroDF[['year', 'quarter', 'realgdp', 'realcons']].set_index(['year', 'quarter'])
macroDF3 = macroDF[['year', 'quarter', 'cpi', 'realint']].set_index(['year', 'quarter'])

print('\n Split dataframe: \n', macroDF2)
print('\n Split dataframe: \n', macroDF3)

#join via year and quarter fields
macroJoin = macroDF2.join(macroDF3)
print('\n Joined dataframe: \n', macroJoin)
```

Provide an example for concat() with index enabled - 1 Point

```python
#split the dataframe to demonstrate how to use concat
macroDF4 = macroDF[['year', 'quarter', 'realgdp']]
macroDF5 = macroDF[['year', 'quarter', 'realcons']]

print('\n Split dataframe: \n', macroDF4)
print('\n Split dataframe: \n', macroDF5)

#concat via columns
macroConcat = pd.concat([macroDF4, macroDF5], axis = 1)
print('\n Concatenated dataframe: \n', macroConcat)
```

Provide an example for concat() with index disabled - 1 Point

```python
#split the dataframe to demonstrate how to use concat
macroDF4 = macroDF[['year', 'quarter', 'realgdp']]
macroDF5 = macroDF[['year', 'quarter', 'realcons']]

#concat via columns (axis = 1), with index disabled
macroConcatNoIndex = pd.concat([macroDF4, macroDF5], axis = 1, ignore_index = True)

print('Concatenated, index disabled, dataframe:\n', macroConcatNoIndex)
```

Create a pivot table - 1 Point

```python
#make pt that finds the average realinv per year
macroPT = macroDF.pivot_table(index = 'year', values = 'realinv', aggfunc = 'mean')

print('\n Pivot table - Average realinv per year: \n', macroPT)
```

Provide an example for melt() - 1 Point

```python
#melt dataframe, make fixed fields (year and quarter) and unpivot realgdp and realcons
macroMelt = macroDF.melt(id_vars = ['year', 'quarter'], value_vars = ['realgdp', 'realcons'], var_name = 'Factors', value_name = 'Value')

print('\n Melt: \n', macroMelt)
```
